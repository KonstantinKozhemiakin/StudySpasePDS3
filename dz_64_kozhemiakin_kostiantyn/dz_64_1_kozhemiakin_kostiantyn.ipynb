```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img1, img2 = cv2.imread('chess_1.png'), cv2.imread('chess_2.png')
```

```python
fig, axis = plt.subplots(1, 2)
axis[0].imshow(img1)
axis[1].imshow(img2)
```

```python
pts_src = np.array([[213, 218], [209, 79], [446, 72],[443, 230]])
pts_dst = np.array([[381, 347],[221, 312],[217, 67],[471, 115]])
```

```python
h, status = cv2.findHomography(pts_dst, pts_src)
```

```python
img3 = cv2.warpPerspective(img2, h, (img2.shape[1], img2.shape[0]))
```

```python
plt.imshow(img3)
```
