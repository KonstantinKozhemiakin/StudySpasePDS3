```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.colors as clr
```

```python
im = cv2.imread("cat.jpg")
plt.axis('off')
plt.imshow(im)
plt.show()
```

```python
R = im.copy()
R[:,:,(1,2)] = 0 
plt.imshow(R)
plt.show()
```

```python
G = im.copy()
G[:,:,(0,2)] = 0
plt.imshow(G)
plt.show()
```

```python
B = im.copy()
B[:,:,(0,1)] = 0
plt.imshow(B)
plt.show()
```

```python
img_HSV = cv2.cvtColor(im, cv2.COLOR_BGR2HSV)
```

```python
H = img_HSV.copy()
H[:, :, (1,2)] = 255
H_RGB =  cv2.cvtColor(H,cv2.COLOR_HSV2RGB) 
plt.show()
plt.imshow(H_RGB)
```

```python
S = img_HSV.copy()
S[:, :, 0] = 179 
S[:, :, 2] = 255 
S_RGB =  cv2.cvtColor(S,cv2.COLOR_HSV2RGB)
plt.show()
plt.imshow(S_RGB)
```

```python
V = img_HSV.copy()
V[:, :, 0] = 179
V[:, :, 1] = 255
V_RGB =  cv2.cvtColor(V,cv2.COLOR_HSV2RGB) 
plt.show()
plt.imshow(V_RGB)
```

```python
im_Lab = cv2.cvtColor(im, cv2.COLOR_BGR2LAB)
```

```python
l = im_Lab.copy()
l[:, :, 1] = 0
l[:, :, 2] = 0
l_RGB =  cv2.cvtColor(l,cv2.COLOR_LAB2RGB)
plt.imshow(l_RGB) 
plt.show()
```

```python
a = im_Lab.copy()
a[:, :, 0] = 0
a[:, :, 2] = 0
a_RGB =  cv2.cvtColor(a,cv2.COLOR_LAB2RGB)
plt.imshow(a_RGB) 
plt.show()
```
