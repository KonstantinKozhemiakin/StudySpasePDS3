```python
print('''1.
Кількість незношених елементів: 4
Кількість зношених елементів: 3
Кількість елементів, які включені: 3
''') 

# загальна кількість способів вибрати 3 елементи з 7
total_combinations = 7 * 6 * 5 // (3 * 2 * 1)

# кількість способів вибрати 2 незношених елементів і 1 зношений
successful_combinations = 4 * 3 * 3 // (2 * 1)

# обчислення ймовірності
P = successful_combinations / total_combinations

print("Ймовірність того, що будуть включені два незношені елементи:", P)
```

```python
print('''2.
Встановлено незалежно діючі системи сигналізації : 2
Ймовірність того, що перша сигналізація спрацює : 0,95
Ймовірність того, що перша сигналізація спрацює : 0,9
''')
print('''А) Ймовірність, що спрацює принаймні один сигналізатор:
1 - P(обидва сигналізатори не спрацюють) 
1 - P(перший сигналізатор не спрацює) * P(другий сигналізатор не спрацює)
''') 
p_at_least_one = 1 - (1 - 0.95) * (1 - 0.9)

print("Ймовірність, що принаймні один сигналізатор спрацює: ", p_at_least_one)

print()

print('''В) Ймовірність, що спрацює тільки один сигналізатор:
P(перший сигналізатор спрацює) * P(другий сигналізатор не спрацює) + P(перший сигналізатор не спрацює) * P(другий сигналізатор спрацює)
''')
p_only_one = 0.95 * (1 - 0.9) + (1 - 0.95) * 0.9
print("Ймовірність, що тільки один сигнальний пристрій спрацює: ", p_only_one)

```

```python
print('''Верстати які виготовляють продукцію на один конвеєр : 2
Перший верстат виробляє виробів вищого гатунку : 60% 
Другий верстат виробляє виробів вищого гатунку : 84%
''')

print('''А) Ймовірність отримати виріб вищого гатунку:
P(вищий гатунок | перший верстат) * P(перший верстат) + P(вищий гатунок | другий верстат) * P(другий верстат)
''')
p_high = 0.6 * 0.5 + 0.84 * 0.5
print("Ймовірність отримати виріб вищого гатунку:", p_high)

print()

print('''В) Ймовірність того, що виріб зроблений на першому верстаті, за умови, що він вищого гатунку:
P(вищий гатунок | перший верстат) * P(перший верстат) / P(вищий гатунок)
''')
p_first_given_high = 0.6 * 0.5 / p_high
print("Ймовірність того, що виріб зроблений на першому верстаті, за умови, що він вищого гатунку:", p_first_given_high)
```
