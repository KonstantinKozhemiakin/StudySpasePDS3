```python
print('''Для вирішення цієї задачі спочатку потрібно ввести стандартизовану випадкову величину. 
Нехай X - кількість телевізорів, які витримали гарантійний термін експлуатації, 
тоді ймовірність того, що телевізор витримає гарантійний термін становить p=0.9.

Оскільки розподіл біноміальний, то середнє значення і дисперсія стандартизованої 
випадкової величини можна знайти за формулами:

M(X) = np = 100*0.9 = 90
D(X) = npq = 100*0.9*0.1 = 9
де n = 100 - загальна кількість телевізорів.

Далі, стандартизуємо випадкову величину:

Z = (X - M(X)) / sqrt(D(X)) 

Z = (85 - 90) / sqrt(9) = -1.67
Z = (95 - 90) / sqrt(9) = 1.67


За таблицею Лапласа, P(Z < -1.67) = 0.0475, P(Z < 1.67) = 0.9525.

Отже, ймовірність того, що кількість телевізорів, які витримали гарантійний термін експлуатації буде в межах [85;95], складе:

P(-1.67 < Z < 1.67) = P(Z < 1.67) - P(Z < -1.67) = 0.9525 - 0.0475 = 0.905
''')


```
