```python
print('''1.
n: 20 
xi = 1, 2, 4, 6, 7
ni = 2, 3, 10, 3, 2
''')
avg = (1*2 + 2*3 + 4*10 + 6*3 + 7*2)/ 20
print("Вибіркове середнє :", avg)
d_x = (((1-avg)**2)*2 + ((2-avg)**2)*3 + ((4-avg)**2)*10 + ((6-avg)**2)*3 + ((7-avg)**2)*2)/(20-1)
print("Виправлена дисперсія :", d_x)
s = d_x**0.5
print("Виправлене середнє квадратичне відхилення :", s)
print("Із таблиці, для n = 20 і гама = 0,95 -> ty = 2,093")
ty = 2.093
a = ty * s / 20**0.5
print("Точність оцінки :", a)
print ("Довірчий інтервал ", avg-a, ":", avg+a)
```
