```python
print('''1.
Кількысть деталей : 3
Ймовірність того, що при перевірці деталь виявиться стандартною : 0.6
Ймовірність того, що при перевірці деталь виявиться бракованою  : 1-0.6 = 0.4
''')
print('''
Ймовірність того, що буде x бракованих деталей з трьох:
P(X = x) = C(3, x) * p^x * q^(3-x), 
де C(3, x) - кількість способів вибрати x елементів з 3, тобто число поєднань з трьох елементів, рівне 3!/(x!*(3-x)!).

Закон розподілу системи двох дискретних випадкових величин:

P(X = 0) = C(3, 0) * p^0 * q^3 = (3! / (0! * (3 - 0)!)) * p^0 * q^3
P(X = 1) = C(3, 1) * p^1 * q^2 = (3! / (1! * (3 - 1)!)) * p^1 * q^2
P(X = 2) = C(3, 2) * p^2 * q^1 = (3! / (2! * (3 - 2)!)) * p^2 * q^1
P(X = 3) = C(3, 3) * p^3 * q^0 = (3! / (3! * (3 - 3)!)) * p^3 * q^0
''') 
p_x_0 = 1 * 0.4**0 * 0.6**3
p_x_1 = 3 * 0.4**1 * 0.6**2
p_x_2 = 3 * 0.4**2 * 0.6**1
p_x_3 = 1 * 0.4**3 * 0.6**0
print("P(X = 0) = P(Y = 3) =", p_x_0) 
print("P(X = 1) = P(Y = 2) =", p_x_1)
print("P(X = 2) = P(Y = 1) =", p_x_2)
print("P(X = 3) = P(Y = 0) =", p_x_3)
print()
mat_exp_x = 0 * p_x_0 + 1 * p_x_1 + 2 * p_x_2 + 3 * p_x_3
mat_exp_y = 0 * p_x_3 + 1 * p_x_2 + 2 * p_x_1 + 3 * p_x_0
print("Математичне очікування бракованої деталі :", mat_exp_x)
print("Математичне очікування стандартної :", mat_exp_y)
print()
disp_x = 0**2 * p_x_0 + 1**2 * p_x_1 + 2**2 * p_x_2 + 3**2 * p_x_3 - mat_exp_x**2
disp_y = (0**2) * p_x_3 + (1**2) * p_x_2 + (2**2) * p_x_1 + (3**2) * p_x_0 - mat_exp_y**2
print("Дисперсія можливості отримання бракованої деталі :", disp_x)
print("Дисперсія можливості отримання стандартної деталі :", disp_y)
print()
sigma_x = disp_x**0.5
sigma_y = disp_y**0.5
print("Стандартне відхилення бракованої деталі :", sigma_x)
print("Стандартне відхилення стандартної деталі :", sigma_y)
print()
mat_exp_x_y = 0*3*p_x_0 + 1*2*p_x_1 + 2*1*p_x_2 + 3*0*p_x_3
print("Умовне математичне очікування XY :", mat_exp_x_y)
print()
cov_x_y = mat_exp_x_y - (mat_exp_x * mat_exp_y)
print("Коефіцієнт коваріації :", cov_x_y)
print()
r_x_y = cov_x_y / (sigma_x * sigma_y)
print("Коефіцієнт кореляції :", r_x_y)
```
