```python
print('''1.
Економісти, які вкладають свої прогнози в одну папку : 2
Ймовірність критичної похибки першого економіста в розрахунках  : 0.1
Ймовірність критичної похибки другого економіста в розрахунках  : 0.2
Перший економіст заповнив документів: 40
Другий економіст заповнив документів: 60
''')
print('''Ймовірність того, що взята навмання папка з критичною похибкою є прогнозом першого аналітика:
P(перший аналітик) * Р(критична похибка | перший аналітик) / Р(критична похибка)
''') 
p_error_first_analytics =  (40/(40+60)) * ((40*0.1)/40) / ((40*0.1)+(60*0.2))/(40+60)

print("Ймовірність того, що взята навмання папка з критичною похибкою є прогнозом першого аналітика: ", p_error_first_analytics)
```

```python
print('''2.
Х: 1000, 1500, 2000
P: 0.5,  0.3 ,  0.2

Y: 1000, 1500, 1750
P: 0.4,  0.4 ,  0.2
''')

mat_x = 1000 * 0.5 + 1500 * 0.3 + 2000 * 0.2
mat_y = 1000 * 0.4 + 1500 * 0.4 + 1750 * 0.2

disp_x = (1000**2 * 0.5) + (1500**2 * 0.3) + (2000**2 * 0.2) - mat_x**2
disp_y = (1000**2 * 0.4) + (1500**2 * 0.4) + (1750**2 * 0.2) - mat_y**2

print("Математичне очікування M(X): ", mat_x)
print("Математичне очікування M(Y): ", mat_y)
print("Дисперсія D(X): ", disp_x)
print("Дисперсія D(Y): ", disp_y)
print()
print('Отже, за критерієм очікуваного прибутку значення однакові, але дисперсія у варіанта "У" менша, відповідно і ризик збитків менший')
```
