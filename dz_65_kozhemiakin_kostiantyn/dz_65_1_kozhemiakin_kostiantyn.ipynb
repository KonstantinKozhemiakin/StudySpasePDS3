```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
img = cv2.imread("titanic.jpg")
plt.imshow(img)
plt.show
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
faces
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 5)
```

```python
plt.imshow(img)
plt.show
```

```python
img_rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotated)
plt.show
```

```python
gray_rotaded = cv2.cvtColor(img_rotated, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray_rotaded, 1.3, 5)
for (x, y, w, h) in faces:
    cv2.rectangle(img_rotated, (x, y), (x+w, y+h), (255, 0, 0), 5)
plt.imshow(img_rotated)
plt.show
```

```python
import dlib
from imutils import face_utils
```

```python
img = cv2.imread("titanic.jpg")
```

```python
gray = cv2.imread('titanic.jpg', 0)
im = np.float32(gray) / 255.0
# Computing gradients
gx = cv2.Sobel(im, cv2.CV_32F, 1, 0, ksize=1)
gy = cv2.Sobel(im, cv2.CV_32F, 0, 1, ksize=1)
mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
```

```python
plt.imshow(mag)
plt.show()
```

```python
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(gray, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(gray, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
gray_rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
face_detect = dlib.get_frontal_face_detector()
rects = face_detect(gray_rotated, 1)

for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(gray_rotated, (x, y), (x + w, y + h), (255, 255, 255), 3)
    
plt.figure(figsize=(10,6))
plt.imshow(gray_rotated, cmap='gray')
plt.show()
```
