```python
from PIL import Image
from PIL import ImageEnhance

class ImageWorker:
    def __init__(self, path):
        self.image = Image.open(path)
    
    def show_image(self):
        display(self.image)
    
    def save_image(self, path):
        self.image.save(path)
    
    def change_contrast(self, contrast_factor, brightness_factor):
        converter = ImageEnhance.Contrast(self.image)
        self.image = converter.enhance(contrast_factor)
        converter = ImageEnhance.Brightness(self.image)
        self.image = converter.enhance(brightness_factor)
    
    def change_brightness(self, brightness_factor):
        converter = ImageEnhance.Brightness(self.image)
        self.image = converter.enhance(brightness_factor)
    
    def change_saturation(self, saturation_factor):
        converter = ImageEnhance.Color(self.image)
        self.image = converter.enhance(saturation_factor)
    
    def resize_image(self, new_width, new_height):
        self.image = self.image.resize((new_width, new_height))
    
    def crop_image(self, left, upper, right, lower):
        self.image = self.image.crop((left, upper, right, lower))
    
    def rotate_image(self, angle):
        self.image = self.image.rotate(angle)
```

```python
my_picture = ImageWorker('cat.jpg')
my_picture.resize_image(800,600)
my_picture.show_image()
```

```python
my_picture = ImageWorker('cat.jpg')
my_picture.resize_image(800,600)
my_picture.change_brightness(1.5)
my_picture.show_image()
```

```python
my_picture.change_contrast(1.5, 1.5)
my_picture.show_image()
```

```python
my_picture.change_saturation(0.0)
my_picture.show_image()
```

```python
my_picture.rotate_image(-90)
my_picture.show_image()
```

```python
my_picture = ImageWorker('cat.jpg')
my_picture.resize_image(800,600)
my_picture.change_contrast(1.5, 1.5)
my_picture.change_saturation(0.0)
my_picture.rotate_image(-90)
width, height = my_picture.image.size
print(width, height)
my_picture.crop_image(200, 0, 600, height)
my_picture.show_image()
```
